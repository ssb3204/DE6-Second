```python
%%sql

DROP TABLE IF EXISTS analytics.monthly_sales_by_industry;
```

```python
%%sql

CREATE TABLE IF NOT EXISTS analytics.monthly_sales_by_industry(
    month VARCHAR(100) NOT NULL,
    card_tpbuz_nm_1 VARCHAR(100) NOT NULL,
    card_tpbuz_nm_2 VARCHAR(100) NOT NULL,
    total_amt BIGINT NOT NULL,
    total_cnt INT NOT NULL,
    avg_unit_price NUMERIC(14,2)
);
```

```python
%%sql

INSERT INTO analytics.monthly_sales_by_industry
SELECT 
    TO_CHAR(ta_ymd, 'YYYY-MM') AS month,
    card_tpbuz_nm_1,
    card_tpbuz_nm_2,
    SUM(amt) AS total_amt,
    SUM(cnt) AS total_cnt,
    CASE WHEN SUM(CNT) = 0 THEN 0 ELSE ROUND(SUM(amt)::FLOAT / SUM(cnt), 2)END AS avg_unit_price
FROM raw_data.card_transaction
GROUP BY 1,2,3
ORDER By 1,2,3;
```

```python
%%sql

SELECT *
FROM analytics.monthly_sales_by_industry
ORDER BY 1,2,3
LIMIT 10;
```

```python
%%sql

SELECT DISTINCT month FROM analytics.monthly_sales_by_industry;
```

```python
%%sql
SELECT *
FROM analytics.monthly_sales_by_industry
WHERE month in ('2024-03')
ORDER BY 1,2,3;
```
