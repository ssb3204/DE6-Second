## RawData 생성 및 조회


```python
%%sql
CREATE TABLE raw_data.card_transaction (
    ta_ymd DATE,
    card_tpbuz_nm_1 VARCHAR(100),
    card_tpbuz_nm_2 VARCHAR(100),
    hour INT,
    sex VARCHAR(1),
    age INT,
    day INT,
    amt INT,
    cnt INT
);
```

## 가공 테이블 생성


## **최근 1년간 남녀별 업종 소비량** [2025.02 ~ 2024.03]

```python
%%sql

CREATE TABLE analytics.gender_industry_spending_stats AS
SELECT
    sex,
    card_tpbuz_nm_1,
    card_tpbuz_nm_2,
    SUM(amt) AS total_amt,
    SUM(cnt) AS total_cnt,
    ROUND(SUM(amt)::FLOAT / NULLIF(SUM(cnt), 0), 2) AS avg_amt_per_txn
FROM
    raw_data.card_transaction
GROUP BY
    sex, card_tpbuz_nm_1, card_tpbuz_nm_2;
```
